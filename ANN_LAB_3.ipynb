```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AND_gate(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  weighted_sum = w1 * x1 + w2 * x2
  # Element-wise comparison using broadcasting
  return np.where(weighted_sum <= theta, 0, 1)  # Use np.where for element-wise assignment

def OR_gate(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.3
  weighted_sum = w1 * x1 + w2 * x2
  # Element-wise comparison using broadcasting
  return np.where(weighted_sum <= theta, 0, 1)  # Use np.where for element-wise assignment

def NOT_gate(x):
  # Return 1 for 0 and 0 for anything else (including non-zero values)
  return np.where(x == 0, 1, 0)  # Element-wise comparison

def XOR_gate(x1, x2):
  s1 = OR_gate(x1, x2)
  s2 = AND_gate(x1, x2)
  return NOT_gate(OR_gate(1 - s1, s2))  # Apply NOT gate to the output of OR(inverted s1, s2)
```

```python
# Generate input values
x_range = (0, 1)
y_range = (0, 1)
x_values = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Calculate outputs for each gate
AND_output = np.array([AND_gate(x[0], x[1]) for x in x_values])
OR_output = np.array([OR_gate(x[0], x[1]) for x in x_values])
XOR_output = np.array([XOR_gate(x[0], x[1]) for x in x_values])
```

```python
print("AND_output: \n",AND_output)
print("\nOR_output: \n",OR_output)
print("\nXOR_output: \n",XOR_output)
```

```python
def plot_decision_boundary(gate_func, x_range, y_range, label):
  """Plots the decision boundary of a given logic gate."""
  x, y = np.meshgrid(x_range, y_range)
  z = gate_func(x, y)

  # Create separate contours for 0 and 1 outputs
  plt.contour(x, y, z, levels=[0.5], colors='black', label='Decision Boundary')
  plt.contourf(x, y, z, levels=[0, 0.5, 1], colors=['blue', 'white', 'red'])

  # Adjust colorbar labels
  plt.colorbar(label='Output')

  plt.xlabel('Input 1 (x1)')
  plt.ylabel('Input 2 (x2)')
  plt.title(f"{label} Decision Boundary")
  plt.xlim(x_range[0], x_range[1])
  plt.ylim(y_range[0], y_range[1])
  plt.legend()  # Add legend for decision boundary and colored regions

  # Add labels for the colored regions (0 and 1)
  levels = [0, 0.5, 1]
  colors = ['blue', 'white', 'red']
  for i, level in enumerate(levels):
    if i == 1:  # Skip white (empty space between contours)
      continue
    plt.text(
        x.max() - 0.1,  # Place labels near the edge
        y.min() + 0.1 * i,
        colors[i],
        ha='right',
        va='bottom',
        fontsize=10,
        weight='bold',
        bbox=dict(boxstyle="round", facecolor=colors[i], alpha=0.2),
        zorder=2  # Ensure labels are on top of contours
    )
```

```python
# Plot decision boundaries
plot_decision_boundary(AND_gate, x_range, y_range, "AND")
plt.show()
```

```python
plot_decision_boundary(OR_gate, x_range, y_range, "OR")
plt.show()
```

```python
plot_decision_boundary(XOR_gate, x_range, y_range, "XOR")
plt.show()
```
