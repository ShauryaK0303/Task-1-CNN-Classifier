```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Activation:
  def __init__(self):
    pass

  def forward(self, X):
    raise NotImplementedError

  def backward(self, output):
    raise NotImplementedError
```

```python
class Sigmoid(Activation):
  def __init__(self):
    super().__init__()

  def forward(self, X):
    return 1 / (1 + np.exp(-X))

  def backward(self, output):
    return output * (1 - output)
```

```python
class ReLU(Activation):
  def __init__(self):
    super().__init__()

  def forward(self, X):
    return np.maximum(0, X)

  def backward(self, output):
    return np.where(output > 0, 1, 0)
```

```python
class Tanh(Activation):
  def __init__(self):
    super().__init__()

  def forward(self, X):
    return np.tanh(X)

  def backward(self, output):
    return 1 - (output * output)
```

```python
class HardLimiter(Activation):
  def __init__(self, threshold):
    super().__init__()
    self.threshold = threshold

  def forward(self, X):
    return np.where(X > self.threshold, self.threshold, 0)

  def backward(self, output):
    # Hard limiter has zero gradient everywhere (derivative is 0)
    return np.zeros_like(output)
```

```python
# Define a function to plot all activation functions
def plot_activations(X_min, X_max):
  X = np.linspace(X_min, X_max, 100)  # Create 100 points between X_min and X_max

  sigmoid = Sigmoid()
  relu = ReLU()
  tanh = Tanh()
  hard_limiter = HardLimiter(threshold=0.5)

  sigmoid_output = sigmoid.forward(X)
  relu_output = relu.forward(X)
  tanh_output = tanh.forward(X)
  hard_limiter_output = hard_limiter.forward(X)

  plt.plot(X, sigmoid_output, label="Sigmoid")
  plt.plot(X, relu_output, label="ReLU")
  plt.plot(X, tanh_output, label="Tanh")
  plt.plot(X, hard_limiter_output, label="Hard Limiter (threshold=0.5)")
  plt.xlabel("Input (X)")
  plt.ylabel("Output")
  plt.title("Different Activation Functions")
  plt.grid(True)
  plt.legend()
```

```python
# Example usage
X_min = -1
X_max = 2

plot_activations(X_min, X_max)
plt.show()
```
