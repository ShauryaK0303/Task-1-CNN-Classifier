```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample data for bar graph
data = {'category': ['A', 'B', 'C', 'D'], 'value': [5, 3, 7, 2]}
df = pd.DataFrame(data)

# Create bar graph
plt.figure(figsize=(8, 6))
plt.bar(df['category'], df['value'], color='orange')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Graph')
plt.show()
```

```python
# Sample data for line graph
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# Create line graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')
plt.show()
```

```python
# Sample data for scatter plot
x = [1, 3, 5, 7, 9]
y = [2, 5, 7, 2, 8]

# Create scatter plot
plt.scatter(x, y, color='red', marker='x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# Sample data for histogram
data = np.random.randn(100)  # Generate random normal distribution data

# Create histogram
plt.hist(data, bins=10, color='red', edgecolor='black')  # Adjust bins for better visualization
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Sample data for pie chart
categories = ['A', 'B', 'C']
values = [30, 45, 25]

# Create pie chart
plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=140)  # Customize options
plt.title('Pie Chart')
plt.show()
```

# Conclusion:
**Bar Graph:**
1. Compare the values between categories. See which category has the highest, lowest, or similar values.
2. Identify patterns or trends across categories.
**Line Graph:**
1. Observe the increase, decrease, or stability of the variable over time (if x-axis represents time).
2. Identify potential relationships between the plotted variable and time.
**Scatter Plot:**
1. Look for any clustering or spread of points, suggesting a positive, negative, or no correlation.
2. Identify outliers that deviate significantly from the overall trend.
**Histogram:**
1. Understand how the data is distributed (centered, skewed, etc.).
2. See if there are any concentration points or gaps in the data.
**Pie Chart:**
1. Determine the relative proportion of each category compared to the whole.
2. Identify the dominant category or compare the sizes of different slices.
